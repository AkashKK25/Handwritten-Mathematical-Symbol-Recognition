```python
# Importing Libraries
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
```

```python
# Loading Data
data_train = np.load('data_train.npy')
labels_train = np.load('t_train.npy')
test = np.load('t_train.npy')

# Data Shapes
print(data_train.shape, labels_train.shape)
print(data_train)

# Pre-processing the data
data_train = data_train.T

# Data Shapes
print(data_train.shape, labels_train.shape)
print(data_train[0])
#print(labels_train)
```

```python
# Ease for edit dictionaries
dictout = {-1:'u', 0:'x', 1:'sqrt', 2:'+', 3:'-', 4:'=', 5:'%', 6:'do', 7:'PI', 8:'pi', 9:'sum'}
dictin = {'u':-1, 'x':0, 'sq':1, '+':2, '-':3, '=':4, '%':5, 'd':6, 'P':7, 'p':8, 'sm':9}

# labels_train[300:400]
```

```python
# Relabelling wrong labels
for i in range(9000,9032):
    plt.imshow(data_train[i].reshape(300,300), cmap = 'gray')
    plt.show()
    print(dictout[labels_train[i]],'\t',i)
    l = input('Enter x,sq,+/-/=/%/d/P/p/sm/u :  ')
    if l == '':
        pass
    else:
        labels_train[i] = dictin[l]
```
